```python
import os
import shutil
import pandas as pd
```

```python
# Pfad zum Ordner mit den Bildern
bildordner = 'C:/Users/Lorena/Documents/Studium/Masterarbeit/Data/images/'

# Pfad zum Ordner, in dem die Bilder gespeichert werden sollen
zielordner = 'C:/Users/Lorena/Documents/Studium/Masterarbeit/Ausarbeitung_files/Final/Data/Images_noDuplicates/'

file_path = "C:/Users/Lorena/Documents/Studium/Masterarbeit/Ausarbeitung_files/Final/Data/HAM10000_metadata.csv"
```

```python
df = pd.read_csv(file_path,  sep=';')

df_unique = df.drop_duplicates(subset='lesion_id', keep='first')
```

```python
df_unique
```

```python
# Liste der einzigartigen Bezeichnungen
bezeichnungen = df_unique['image_id']

# Schleife über die Bezeichnungen
for bezeichnung in bezeichnungen:
    # Schleife über die Dateien im Bildordner
    for dateiname in os.listdir(bildordner):
        if bezeichnung in dateiname:
            # Pfade zu den Quelldateien und Zieldateien erstellen
            quelle = os.path.join(bildordner, dateiname)
            ziel = os.path.join(zielordner, dateiname)
            
            # Bild kopieren oder verschieben
            shutil.copy(quelle, ziel)  # oder shutil.move(quelle, ziel), wenn Sie die Dateien verschieben möchten
```
